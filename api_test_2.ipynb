```python
import requests
import json
import pandas as pd
```

```python
base_url = 'https://api.artic.edu/api/v1/artworks'
```

```python
requests.get(base_url).json()
```

```python
total_pages = requests.get(base_url + '?page=' + '0').json()['pagination']['total_pages']
title = []
image_id = []
medium_display = []
for x in range(40):
    print(x, '/', total_pages)
    try:
        works = requests.get(base_url + '?page=' + str(x)).json()['pagination']['limit']
        for y in range(works):
            title.append(requests.get(base_url + '?page=' + str(x)).json()['data'][y]['title'])
            image_id.append(requests.get(base_url + '?page=' + str(x)).json()['data'][y]['image_id'])
            medium_display.append(requests.get(base_url + '?page=' + str(x)).json()['data'][y]['medium_display'])
    except:
        continue
```

```python
pd.DataFrame({'title': title, 'image_id': image_id, 'medium_display': medium_display})
```
